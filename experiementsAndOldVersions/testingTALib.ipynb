```python
import yfinance as yf
import talib as ta
import matplotlib.pyplot as plt

# Get stock data from AAPL
aapl = yf.Ticker("AAPL")
data = aapl.history(period="1y")

# Calculate RSI using talib
close = data['Close'].values
rsi = ta.RSI(close)

# Plotting the stock closing price
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(data.index, data['Close'])
plt.title('AAPL Stock Closing Price')
plt.xlabel('Date')
plt.ylabel('Price')

# Plotting the RSI
plt.subplot(2, 1, 2)
plt.plot(data.index, rsi)
plt.title('Relative Strength Index (RSI)')
plt.xlabel('Date')
plt.ylabel('RSI')

# Adjusting the layout
plt.tight_layout()

# Display the plot
plt.show()
```
