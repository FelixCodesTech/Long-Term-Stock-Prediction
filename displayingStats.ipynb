# Imports

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

# Preparing Data

```python
data = pd.read_csv('statistics.csv', index_col=0)

# We're gonna color the points based on how good the trainMSE is
# The more Red the worse (0 min, 1 max), the more Green the better (0 min, 1 max)

minTrainMSE = data["trainMSE"].min()
maxTrainMSE = data["trainMSE"].max()

data["Red"] = (data["trainMSE"] - minTrainMSE) / (maxTrainMSE - minTrainMSE)
data["Green"] = 1 - data["Red"]
```

```python
data.describe()
```

# Plotting

```python
plt.figure(figsize=(20, 5))
plt.scatter(data["solutionWindow"], data["contextWindow"], c=data["Red"], cmap='jet')
plt.colorbar()
plt.xlabel("Solution Window")
plt.ylabel("Context Window")
plt.title("Context Window vs Solution Window")

# Add more indicators on x-axis
plt.xticks(range(0, 1050, 20), rotation=-40)  # Set label locations.


plt.show()
```
